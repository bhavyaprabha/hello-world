```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv('TrainingData.csv', header=0)
```

```python
chunk_ids=dataset.chunkID.unique()
```

```python
chunks = dict()
# sort rows by chunk id
for chunk_id in chunk_ids:
    selection=dataset[dataset.chunkID==chunk_id]
    chunks[chunk_id]=selection
#total chunks 208
```
