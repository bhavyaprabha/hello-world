```python
import urllib.request
```

```python
#specify the url
wiki = "http://10.41.72.110:8000/scoreboard_test/"
```

```python
#Query the website and return the html to the variable 'page'
page = urllib.request.urlopen(wiki)
```

```python
page
```

```python
#import the Beautiful soup functions to parse the data returned from the website
from bs4 import BeautifulSoup
```

```python
#Parse the html in the 'page' variable, and store it in Beautiful Soup format
soup = BeautifulSoup(page)
```

```python
soup.prettify()
```

```python
#we can work with html tags 
soup.title.string
```

```python
#to get all components in div
table=soup.find_all('table',class_='table table-striped table-hover')
```

```python
#we can get all parameters with in one tag like href,noframes...
pa=soup.find_all('table')
for i in pa:
    print(i)
```

```python
#Generate lists
A=[]
B=[]
C=[]
D=[]
E=[]
for row in table:
    s=row.find_all('tr')
for row in s:
    line=row.find_all('td')
    if (len(line)==5):
        A.append(line[0].find(text=True))
        B.append(line[1].find(text=True))
        C.append(line[2].find(text=True))
        D.append(line[3].find(text=True))
        E.append(line[4].find(text=True))
```

```python
#to get column headers from the table
head=s[0].find_all('th')
columns=[]
for i in head:
    columns.append(i.string)
```

```python
col=columns
```

```python
import pandas as pd
df=pd.DataFrame(A)
df[1]=B
df[2]=C
df[3]=D
df[4]=E
```

```python
df.columns=col
```
