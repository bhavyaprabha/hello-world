# Write a Python program to test if a given page is found or not on the server.

```python
import urllib.request
import urllib.error
```

```python
url1="https://abcxyz.com/"
url2="https://www.capgemini.com/"
```

```python
try:
    page1=urllib.request.urlopen(url1)
except urllib.error.HTTPError as e:
    print("HTTP Error")
except urllib.error.URLError as e:
    print("URL Error")
else:
    print(page1.read())
    
    
```

```python
try:
    page2=urllib.request.urlopen(url2)
except urllib.error.HTTPError as e:
    print("HTTP Error")
except urllib.error.URLError as e:
    print("URL Error")
else:
    print(page2.read())
```

```python
#gives staus code
page2.getcode()
```

# Download and display the content of robot.txt for en.wikipedia.org

```python
import requests
```

```python
response=requests.get("https://en.wikipedia.org/robots.txt",verify=False)
#verify parameter is used to avoid SSH errors (authentication errors)
```

```python
test = response.text
print("robots.txt for http://www.wikipedia.org/")
print("===================================================")
print(test)
```

# Python Web Scraping: Convert an address into geographic coordinates

```python
geo_url = 'http://maps.googleapis.com/maps/api/geocode/json'
my_address = {'address': '21 Ramkrishana Road, Burdwan, East Burdwan, West Bengal, India', 
             'language': 'en'}
response = requests.get(geo_url, params = my_address)
results = response.json()['results']
print(results)
#my_geo = results[0]['geometry']['location']
#print("Longitude:",my_geo['lng'],"\n","Latitude:",my_geo['lat'])
```
